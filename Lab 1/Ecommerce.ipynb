```python
df['Sales'] = df['Quantity'] * df['UnitPrice']
```

```python
product_sales = df.groupby('Description')['Sales'].sum()
```

```python
top_10_products = product_sales.sort_values(ascending=False).head(10)
```

```python
plt.figure()
plt.bar(top_10_products.index, top_10_products.values)
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['Sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales.index, daily_sales.values)
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Daily Sales Trend")
plt.xticks(rotation=45)
plt.show()
```
